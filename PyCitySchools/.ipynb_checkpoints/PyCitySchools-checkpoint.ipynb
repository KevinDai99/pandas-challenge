```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete.head()
```

```python
# This is to make sure we do not have any duplicates and are dataset is clean
ID_Student = school_data_complete["Student ID"].duplicated().sum()
print(ID_Student)
```

#### Calculating Parameters 

```python
# Total Number of Schools 
total_number_of_schools = len(school_data_complete["School ID"].unique()) ### In case two schools/student
                                                                            # have the same name - use ID
# Total Number of Students
total_number_of_students = school_data_complete["Student ID"].count()

# Total Budget 
# Cannot Sum Budget Column, must groupby school
budget = school_data_complete[["School ID", "budget"]]
budget = budget.groupby("School ID").mean()

total_budget = budget["budget"].sum().astype(int)

# Average Math Score
avg_math_score = school_data_complete["math_score"].mean()

# Average Reading Score 
avg_reading_score = school_data_complete["reading_score"].mean()

# Percentage of students with a passing math score (70 or greater)
pass_math = school_data_complete[school_data_complete["math_score"] >= 70]
percent_math = len(pass_math)/len(school_data_complete)*100

# Percentage of students with a passing eng score (70 or greater)
pass_eng = school_data_complete[school_data_complete["reading_score"] >= 70]
percent_eng = len(pass_eng)/len(school_data_complete)*100

# Percent of students with passing eng and math (70 or greater)
pass_both = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
percent_both = len(pass_both)/len(school_data_complete)*100
```

```python
df_dict = {"Total Schools":total_number_of_schools,
           "Total Students":total_number_of_students ,
           "Total Budget":total_budget,
           "Average Math Score":avg_math_score,
           "Average Reading Score":avg_reading_score,
           "% Passing Math":percent_math,
           "% Passing Reading":percent_eng,
           "% Overall Passing":percent_both}

df = pd.DataFrame(df_dict, index = [0])
```

### Formating DataFrame

```python
# Looping through columns for formating 
columns_1 = ["Total Schools", "Total Students", "Average Math Score", "Average Reading Score"]

for column in columns_1:
    df[column] = df[column].map("{:,.0f}".format)
```

```python
df["Total Budget"] = df["Total Budget"].map("${:,.2f}".format)
```

```python
# Looping through columns for formating 
columns_2 = ["% Passing Math", "% Passing Reading", "% Overall Passing"]

for column in columns_2:
    df[column] = df[column].map("{:.2f}%".format)
    
```

## School Summary 

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math

  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_data_complete.head()
```

```python
groupby = school_data_complete.groupby(["school_name", "type"]).agg(
    {"Student ID":"count", "budget": "mean", "reading_score": "mean",
    "math_score":"mean"})
```

```python
groupby.reset_index(inplace = True)
```

```python
groupby.rename({"type": "School Type",
                "Student ID": "Total Students",
                "budget" : "Total School Budget", 
                "reading_score": "Average Reading Score",
                "math_score": "Average Math Score"}, axis = 1, inplace = True)
```

```python
groupby.set_index("school_name", inplace = True)

# Removing Index Name 
groupby.index.name = None
```

```python
groupby["Per Student Budget"] = groupby["Total School Budget"]/groupby["Total Students"]
```

```python
# Re-organize Columns 
groupby = groupby[['School Type', 'Total Students', 'Total School Budget','Per Student Budget', 
       'Average Reading Score', 'Average Math Score']]
```

```python
groupby.head()
```

### Must groupby and Merge to find % passing students 
----
A single groupby cannot find percent of passing students. To do so, mutiple groupbys are needed. The first groupby finds total students (as seen above), and the anothers the find total students that are over 70% for each reading, math, and both. 

```python
school_data_complete.head()
```

```python
# Filter data to calculate the % pass for each score 
eng_student = school_data_complete[school_data_complete["reading_score"] >= 70]
math_student = school_data_complete[school_data_complete["math_score"] >= 70]
math_eng_student = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
```

```python
# Loop to aggregate data for each filtered dataset 
dataframes = [math_student, eng_student, math_eng_student]
df_group = []

for df in dataframes:
    df = df.groupby("school_name").agg({"Student ID": 'count'})
    df_group.append(df)
```

```python
# rename columns to make df understandable
df_group[0].rename({"Student ID": "% Passing Math"}, axis = 1, inplace = True)
df_group[1].rename({"Student ID": "% Passing Reading"}, axis = 1,inplace = True)
df_group[2].rename({"Student ID": "% Overall Passing"}, axis =1, inplace = True)
```

```python
# Merge all the school % passing parameters 
merged = pd.concat(df_group, axis = 1)
```

```python
merged.head()
```

### Merge with Main DataFrame

```python
df_summary = pd.merge(groupby, merged, how = "left", left_index = True, right_index = True)
```

```python
# Convert the following columns to percent of total students 
columns_divide = ["% Passing Math", "% Passing Reading", "% Overall Passing"]

for column in columns_divide:
    df_summary[column] = df_summary[column]/df_summary["Total Students"]*100
```

```python
# Created Copy, df_summary_format to showcase with formating, df_summary needs to aggregated for further analysis later
df_summary_format = df_summary.copy()
```

#### Formating

```python
df_summary_format["Total School Budget"] = df_summary_format["Total School Budget"].map("${:,.2f}".format)
df_summary_format["Per Student Budget"] = df_summary_format["Per Student Budget"].map("${:,.2f}".format)
```

## Math Scores by Grade 

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
df_grade = school_data_complete[["school_name", "grade", "math_score", "reading_score"]]
df_grade.set_index("school_name", inplace = True)
```

```python
df_grade.head()
```

```python
# creating pd series to be merged 
df_9th_m = df_grade.loc[df_grade["grade"] == '9th', "math_score"].groupby("school_name").mean()
df_10th_m = df_grade.loc[df_grade["grade"] == '10th', "math_score"].groupby("school_name").mean()
df_11th_m = df_grade.loc[df_grade["grade"] == '11th', "math_score"].groupby("school_name").mean()
df_12th_m = df_grade.loc[df_grade["grade"] == '12th', "math_score"].groupby("school_name").mean()
```

```python
# merging the following series 
df_grades_math = [df_9th_m, df_10th_m, df_11th_m, df_12th_m]

df_Math_Score = pd.concat(df_grades_math, axis = 1)

df_Math_Score.columns = ["9th", "10th", "11th", "12th"]
```

```python
# Removing Index Name 
df_Math_Score.index.name = None
```

```python
bins = [0, 585, 630, 645, 680,]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

df_summary["Spending Ranges (Per Student)"] = pd.cut(df_summary["Per Student Budget"], bins = bins, labels = labels)
```
